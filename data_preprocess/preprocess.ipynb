**Data Processing File**

Reading and combining the datasets of the 33 types of cancer into 1

```python
from Data import *

path = "../All_Datasets/"
df = loadDataset(path)
df = mergeDatasets(df)

data = Data(df)
data.df
```

Sorting genes by chromosome names

```python
data.df.sort_index(inplace=True)
data.df
```

Sorting dataset so that labels are in alphabetical order

```python
data.df.sort_values(by="labels",axis=1,inplace=True)
data.df
```

Resetting column indeces

```python
data.df.columns = range(data.df.shape[1])
data.df
```

Separating labels from dataset

```python
data.labels = data.df.loc["labels"]
data.labels
```

```python
data.df = data.df.drop("labels",)
data.df = data.df.astype(float)
data.df
```

Gene scale reduction

```python
data.df = data.df.applymap(lambda x: scaleReduction(x))
data.df
```

Gene noise reduction

```python
data.df = data.df.applymap(lambda x: noiseReduction(x))
data.df
```

Filtering out genes with almost unchanged expression levels using their variance

```python
data.geneFiltering()
data.df
```

Data reshaping in 2D images

```python
all_reshaped = data.reshapeData()
all_reshaped
```

Reading the dimensions of each label in alphabetical order

```python
dimensions = pd.read_csv("../Dataset/dims.csv", sep=',', dtype=int, header=None)
dimensions
```

Printing the first image of each tumor type

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(40, 10))
count=0
pos=1

for i, row in dimensions.iterrows():
    img = all_reshaped[count]
    fig.add_subplot(3, 11, pos)
    plt.title(data.labels[count], fontsize=15)
    plt.imshow(img, interpolation='nearest')

    count += dimensions.iloc[i, 1]
    pos += 1
```

Concatenating all images in one dataset

```python
data.df = pd.DataFrame(np.concatenate(all_reshaped))
data.df
```

Image normalization

```python
data.df = data.df.applymap(lambda x: imageNormalization(x))
data.df
```

Exporting 9285 images in all_data.csv and 9285 labels in labels.csv

```python
data.df.to_csv("../Dataset/input_data.csv")
data.labels.to_csv("../Dataset/labels.csv")
```
